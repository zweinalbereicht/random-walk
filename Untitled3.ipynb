```python
%%time
alpha=1.5
walker=md.ZipfWalker(alpha)
r={}
nbsimus=500
for N in np.logspace(2,4,10):
    print(N)
    r[N]=md.discovery_times_before_cover_time_distribution(0,int(N),walker,nbsimus)
```

```python
md.discovery_times_before_cover_time_distribution()
```

```python
a_values=[0.1,0.5,0.8]
res={}
for a in a_values:
    res[a]={}
    for N in np.logspace(2,4,10):
        tmp=np.array(r[N])
        cover_times=np.max(tmp,axis=1)
        mean_fraction_left=np.mean([len(tmp[i][tmp[i]>a*c])/N for i,c in enumerate(cover_times)])
        res[a][N]=mean_fraction_left
```

```python
for key, value in res.items():
    print(key)
    Ns=np.array(list(value.keys()))
    zs=list(value.values())
    plt.scatter(Ns,zs,label=key)
    plt.plot(Ns,Ns**(-key))

plt.xscale('log')
plt.yscale('log')
```
