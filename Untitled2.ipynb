```python
import ContinuousWalker as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
walker=md.ContinuousWalker()
```

```python
%%time
nbsimus=1000
ls=np.logspace(1,2,20)
ct=[]
for l in ls:
    ct.append(md.cover_time_mean(walker,int(l),nbsimus))
```
