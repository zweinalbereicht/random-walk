```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import DiscreteWalker as md
```

```python
walker=md.DiscreteWalker()
t=np.array(md.territory_distribution_infinite(1,10000,walker,1000000))
#t=np.array(md.fpt_arrival_distribution(1,10000,walker,100000))
t=t[t>0]
```

```python
walker2=md.ZipfWalker(1.3)
t2=np.array(md.territory_distribution_infinite(1,10000,walker2,1000000))
#t2=np.array(md.fpt_arrival_distribution(1,10000,walker2,100000))
t2=t2[t2>0]
```

```python
h=np.histogram(t[t>0], bins=np.linspace(1,1000,500), density=True)
h2=np.histogram(t2[t2>0], bins=np.linspace(1,1000,500), density=True)
```

```python
xs=np.linspace(1,1000)
plt.plot(h[1][1:],h[0], label='brownien')
plt.plot(h2[1][1:],h2[0], label='levy 1.3')
plt.plot(xs,1*xs**(-2))
plt.plot(xs,0.2*xs**(-1.3))
plt.xscale('log')
plt.yscale('log')
plt.legend()
```
