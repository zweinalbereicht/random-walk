**Overshoot check for heavy tailed walkers**

```python
import ContinuousWalker as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
walker=md.AlexWalker()
```

```python
x=1e3
r=md.bounded_overshoot_distribution(0,x,walker,1000000)
```

```python
r=np.array(r)
ls=np.logspace(0,5)
sp=[len(r[r>l])/len(r) for l in ls]
```

```python
fig=plt.figure()
ax,fig=plt.subplots(1,1,figsize=(15,10))

plt.scatter(ls,sp)
plt.plot(ls,ls**(-1/4))
plt.plot(ls,5*ls**(-1/2))
plt.xscale('log')
plt.yscale('log')
plt.title(f'L={x}, Alexwalker, overshoot cumulative')
plt.xlabel('x')
```

**We do observe two regimes, based on the distance of the farthest wall... Still this guy is for sure not integrable, so that's not the right way to go I think for now.**
